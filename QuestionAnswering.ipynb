```python
from transformers import pipeline
```

```python
roberta = pipeline("question-answering", model="modelo utilizado")
```

```python
import pdfplumber

def extractPdf(pdf_path):
    '''
    Método para exetração do conteúdo do pdf
    '''
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text
```

```python
from docx import Document

def extractDocx(path):
    '''
    Método para extração do conteúdo do DOCX
    '''
    doc = Document(path)
    return '\n'.join([para.text for para in doc.paragraphs])
```

```python
pdfText = extractPdf("doencas_respiratorias_cronicas.pdf")
docxText = extractDocx("DICIONARIO_DE_DADOS.docx")
```

```python
def answerQuestion(model, question, context):
    answers = []

    result = roberta(question=question, context=context)
    answers.append({
            "modelo": roberta.model.name_or_path,
            "resposta": result['answer'],
            "pontuação": result['score']
        })

    return answers
```

```python
question = "O que é o aerosol e qual sua composição química ?"

respostas = answerQuestion(roberta, question, pdfText)
for r in respostas:
    print(f"\nModelo: {r['modelo']}")
    print(f"Resposta: {r['resposta']}")
    print(f"Confiança: {r['pontuação']:.2f}")
```
